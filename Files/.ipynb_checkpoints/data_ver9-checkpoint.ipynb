```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVR
from sklearn.multioutput import MultiOutputRegressor
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
dataset = pd.read_csv('Exp_Mn_Mw_Value.txt', sep='\t')
X = dataset.iloc[:, 1:5].values   # columns 1–4: Factor A–D
y = dataset.iloc[:, 5:7].values   # columns 5–6: Responses Mn, Mw
dataset.head()
```

```python
#normalize the data
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_normalized = scaler_X.fit_transform(X)
y_normalized = scaler_y.fit_transform(y)
```

```python
#Split data (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X_normalized, y_normalized, test_size=0.2, random_state=42
)
print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
```

```python
#Neural Network Analysis
def create_neural_network(neurons):
    """Create a neural network with 1 hidden layer (no warnings)"""
    model = Sequential([
        tf.keras.layers.Input(shape=(4,)),  # Input layer (4 features)
        Dense(neurons, activation='relu'),   # 1 hidden layer
        Dense(2)                            # Output layer for 2 responses
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# Test 3 different neuron counts (including 13)
neuron_counts = [8, 13, 20]
nn_results = {}

print("NEURAL NETWORK ANALYSIS")

for neurons in neuron_counts:
    print(f"\n Testing Neural Network with {neurons} neurons...")
    
    # 5-fold cross validation
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    r2_scores = []
    
    fold = 1
    for train_idx, val_idx in kfold.split(X_train):
        X_train_fold, X_val_fold = X_train[train_idx], X_train[val_idx]
        y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]
        
        # Create and train model
        model = create_neural_network(neurons)
        early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
        
        model.fit(X_train_fold, y_train_fold, 
                 validation_data=(X_val_fold, y_val_fold),
                 epochs=200, batch_size=8, verbose=0,
                 callbacks=[early_stopping])
        
        # Predict and calculate R²
        y_pred_fold = model.predict(X_val_fold, verbose=0)
        r2 = r2_score(y_val_fold, y_pred_fold)
        r2_scores.append(r2)
        
        print(f"  Fold {fold}: R² = {r2:.4f}")
        fold += 1
    
    avg_r2 = np.mean(r2_scores)
    std_r2 = np.std(r2_scores)
    nn_results[neurons] = {'avg_r2': avg_r2, 'std_r2': std_r2, 'scores': r2_scores}
    
    print(f"  Average R²: {avg_r2:.4f} ± {std_r2:.4f}")

# Find best neural network configuration
best_neurons = max(nn_results.keys(), key=lambda k: nn_results[k]['avg_r2'])
print(f"\n Best Neural Network: {best_neurons} neurons (R² = {nn_results[best_neurons]['avg_r2']:.4f})")

# Train final neural network with best configuration
final_nn_model = create_neural_network(best_neurons)
early_stopping = EarlyStopping(monitor='loss', patience=20, restore_best_weights=True)
final_nn_model.fit(X_train, y_train, epochs=200, batch_size=8, verbose=0, callbacks=[early_stopping])

# Make predictions on test set
y_pred_nn = final_nn_model.predict(X_test, verbose=0)
nn_r2 = r2_score(y_test, y_pred_nn)

print(f"Final Neural Network Test R²: {nn_r2:.4f}")
```

```python
#Support Vector Machine Analysis
print("\nSUPPORT VECTOR MACHINE ANALYSIS")


# Create SVM model
svm_model = MultiOutputRegressor(SVR(kernel='rbf', C=100, gamma='scale'))

# 5-fold cross validation for SVM
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
svm_r2_scores = []

fold = 1
for train_idx, val_idx in kfold.split(X_train):
    X_train_fold, X_val_fold = X_train[train_idx], X_train[val_idx]
    y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]
    
    # Train SVM
    svm_fold = MultiOutputRegressor(SVR(kernel='rbf', C=100, gamma='scale'))
    svm_fold.fit(X_train_fold, y_train_fold)
    
    # Predict and calculate R²
    y_pred_svm_fold = svm_fold.predict(X_val_fold)
    r2 = r2_score(y_val_fold, y_pred_svm_fold)
    svm_r2_scores.append(r2)
    
    print(f"Fold {fold}: R² = {r2:.4f}")
    fold += 1

avg_svm_r2 = np.mean(svm_r2_scores)
std_svm_r2 = np.std(svm_r2_scores)
print(f"Average SVM R²: {avg_svm_r2:.4f} ± {std_svm_r2:.4f}")

# Train final SVM on full training data
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
svm_r2 = r2_score(y_test, y_pred_svm)
print(f"Final SVM Test R²: {svm_r2:.4f}")

# Convert predictions back to original scale
y_test_original = scaler_y.inverse_transform(y_test)
y_pred_nn_original = scaler_y.inverse_transform(y_pred_nn)
y_pred_svm_original = scaler_y.inverse_transform(y_pred_svm)

print("\nCOMPARISON RESULTS")


# Neural Network Cross-Validation Results
print("\n NEURAL NETWORK CROSS-VALIDATION RESULTS:")
nn_comparison_df = pd.DataFrame({
    'Neurons': neuron_counts,
    'Average R²': [nn_results[n]['avg_r2'] for n in neuron_counts],
    'Std R²': [nn_results[n]['std_r2'] for n in neuron_counts]
})
print(nn_comparison_df.round(4))

# Neural Network Test Predictions
print(f"\n NEURAL NETWORK TEST PREDICTIONS (Best: {best_neurons} neurons):")
nn_results_df = pd.DataFrame({
    'Actual Response 1': y_test_original[:, 0],
    'Predicted Response 1 (NN)': y_pred_nn_original[:, 0],
    'Actual Response 2': y_test_original[:, 1],
    'Predicted Response 2 (NN)': y_pred_nn_original[:, 1]
})
print(nn_results_df.round(2))

# SVM Test Predictions
print(f"\n SVM TEST PREDICTIONS:")
svm_results_df = pd.DataFrame({
    'Actual Response 1': y_test_original[:, 0],
    'Predicted Response 1 (SVM)': y_pred_svm_original[:, 0],
    'Actual Response 2': y_test_original[:, 1],
    'Predicted Response 2 (SVM)': y_pred_svm_original[:, 1]
})
print(svm_results_df.round(2))

# Calculate R² for each response separately
nn_r2_resp1 = r2_score(y_test_original[:, 0], y_pred_nn_original[:, 0])
nn_r2_resp2 = r2_score(y_test_original[:, 1], y_pred_nn_original[:, 1])
svm_r2_resp1 = r2_score(y_test_original[:, 0], y_pred_svm_original[:, 0])
svm_r2_resp2 = r2_score(y_test_original[:, 1], y_pred_svm_original[:, 1])

print(f"\n R² SCORES SUMMARY:")
summary_df = pd.DataFrame({
    'Model': ['Neural Network', 'SVM'],
    'Response 1 R²': [nn_r2_resp1, svm_r2_resp1],
    'Response 2 R²': [nn_r2_resp2, svm_r2_resp2],
    'Overall R²': [nn_r2, svm_r2]
})
print(summary_df.round(4))
```

```python
# CELL 5: Create all the plots
plt.figure(figsize=(15, 10))

# Plot 1: Neural Network Cross-Validation Comparison
plt.subplot(2, 3, 1)
neurons = [str(n) for n in neuron_counts]
avg_scores = [nn_results[n]['avg_r2'] for n in neuron_counts]
std_scores = [nn_results[n]['std_r2'] for n in neuron_counts]
plt.bar(neurons, avg_scores, yerr=std_scores, capsize=5, alpha=0.7, color='blue')
plt.title('Neural Network: R² vs Neurons')
plt.xlabel('Number of Neurons')
plt.ylabel('R² Score')
plt.ylim(0, 1)
for i, v in enumerate(avg_scores):
    plt.text(i, v + 0.05, f'{v:.3f}', ha='center')

# Plot 2: Neural Network Response 1 - Actual vs Predicted
plt.subplot(2, 3, 2)
plt.scatter(y_test_original[:, 0], y_pred_nn_original[:, 0], alpha=0.7, color='blue')
plt.plot([y_test_original[:, 0].min(), y_test_original[:, 0].max()], 
         [y_test_original[:, 0].min(), y_test_original[:, 0].max()], 'r--', lw=2)
plt.xlabel('Actual Response 1')
plt.ylabel('Predicted Response 1')
plt.title(f'Neural Network - Response 1\nR² = {nn_r2_resp1:.3f}')

# Plot 3: Neural Network Response 2 - Actual vs Predicted
plt.subplot(2, 3, 3)
plt.scatter(y_test_original[:, 1], y_pred_nn_original[:, 1], alpha=0.7, color='blue')
plt.plot([y_test_original[:, 1].min(), y_test_original[:, 1].max()], 
         [y_test_original[:, 1].min(), y_test_original[:, 1].max()], 'r--', lw=2)
plt.xlabel('Actual Response 2')
plt.ylabel('Predicted Response 2')
plt.title(f'Neural Network - Response 2\nR² = {nn_r2_resp2:.3f}')

# Plot 4: SVM Response 1 - Actual vs Predicted
plt.subplot(2, 3, 4)
plt.scatter(y_test_original[:, 0], y_pred_svm_original[:, 0], alpha=0.7, color='green')
plt.plot([y_test_original[:, 0].min(), y_test_original[:, 0].max()], 
         [y_test_original[:, 0].min(), y_test_original[:, 0].max()], 'r--', lw=2)
plt.xlabel('Actual Response 1')
plt.ylabel('Predicted Response 1')
plt.title(f'SVM - Response 1\nR² = {svm_r2_resp1:.3f}')

# Plot 5: SVM Response 2 - Actual vs Predicted
plt.subplot(2, 3, 5)
plt.scatter(y_test_original[:, 1], y_pred_svm_original[:, 1], alpha=0.7, color='green')
plt.plot([y_test_original[:, 1].min(), y_test_original[:, 1].max()], 
         [y_test_original[:, 1].min(), y_test_original[:, 1].max()], 'r--', lw=2)
plt.xlabel('Actual Response 2')
plt.ylabel('Predicted Response 2')
plt.title(f'SVM - Response 2\nR² = {svm_r2_resp2:.3f}')

# Plot 6: Model Comparison
plt.subplot(2, 3, 6)
models = ['Neural Network', 'SVM']
overall_r2 = [nn_r2, svm_r2]
colors = ['blue', 'green']
bars = plt.bar(models, overall_r2, color=colors, alpha=0.7)
plt.title('Model Comparison - Overall R²')
plt.ylabel('R² Score')
plt.ylim(0, 1)
for i, v in enumerate(overall_r2):
    plt.text(i, v + 0.02, f'{v:.3f}', ha='center')

plt.tight_layout()
plt.show()

if nn_r2 > svm_r2:
    print(f" Best Neural Network with {best_neurons} neurons (R² = {nn_r2:.4f})")
else:
    print(f"Support Vector Machine (R² = {svm_r2:.4f})")
```
