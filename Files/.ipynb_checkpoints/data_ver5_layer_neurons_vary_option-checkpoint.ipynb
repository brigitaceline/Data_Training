```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping
```

```python
dataset = pd.read_csv('Exp_Mn_Mw_Value.txt', sep='\t')
```

```python
X = dataset.iloc[:, 1:5].values   # columns 1–4: Factor A–D
y = dataset.iloc[:, 5:7].values   # columns 5–6: Responses Mn, Mw
```

```python
dataset.head()
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.1, shuffle=True, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, shuffle=True, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras import Input

model = Sequential()
model.add(Input(shape=(X_train.shape[1],)))  # define the input shape here
model.add(Dense(18, activation='tanh',
                kernel_regularizer=regularizers.l2(0.001)))
model.add(Dense(2))  # 2 outputs: Mn, Mw
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
print(X_train.shape)  # (samples, features)
print(y_train.shape)  # (samples, 2)
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=100, batch_size=4,
                    validation_data=(X_val, y_val), callbacks=[early_stop], shuffle=True)
```

```python
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 :", r2_score(y_test, y_pred))
```

```python
comparison = np.hstack((y_test, y_pred))
print("Actual Mn, Actual Mw, Predicted Mn, Predicted Mw:")
print(comparison)
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Input
from sklearn.metrics import mean_squared_error
```

```python
from tensorflow.keras.optimizers import Adam
# Define transfer functions and architectures
transfer_functions = {
    'tanh': 'Tansig',
    'sigmoid': 'Logsig', 
    'linear': 'Purelin',
    'relu': 'Radbas'
}

architectures = {
    'Single_64': [64],
    'Single_32': [32], 
    'Single_128': [128],
    'Double_64_64': [64, 64],  # Your original
    'Double_32_16': [32, 16],
    'Triple_64_32_16': [64, 32, 16]
}

# Store results
results = []

# Loop over both activation functions and architectures
for activation, tf_name in transfer_functions.items():
    for arch_name, layers in architectures.items():
        
        # Build the model
        model = Sequential()
        model.add(Input(shape=(X_train.shape[1],)))
        
        # Add hidden layers based on architecture
        for neurons in layers:
            model.add(Dense(neurons, activation=activation))
            
        model.add(Dense(2, activation='linear'))  # 2 outputs: Mn & Mw
        
        # Compile the model
        model.compile(optimizer=Adam(), loss='mse')
        
        # Train the model
        model.fit(X_train, y_train,
                 batch_size=32,
                 epochs=25,
                 validation_data=(X_val, y_val),
                 verbose=0,
                 shuffle=True)
        
        # Predict
        y_pred = model.predict(X_test)
        
        # Calculate MSE only for Mw (second column)
        mse_mw = mean_squared_error(y_test[:, 1], y_pred[:, 1])
        
        # Store results
        results.append({
            'activation': tf_name,
            'architecture': arch_name,
            'mse_mw': mse_mw
        })
        
        print(f"Completed: {tf_name} with {arch_name} - MSE: {mse_mw:.4f}")
```

```python
# Convert results to DataFrame for easy viewing
results_df = pd.DataFrame(results)
print("\nAll Results:")
print(results_df.sort_values('mse_mw'))

# Find best combination
best_result = min(results, key=lambda x: x['mse_mw'])
print(f"\nBest combination:")
print(f"Activation: {best_result['activation']}")
print(f"Architecture: {best_result['architecture']}")
print(f"MSE: {best_result['mse_mw']:.4f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results)
print("All Results:")
print(results_df.sort_values('mse_mw'))

# Find best combination
best_result = min(results, key=lambda x: x['mse_mw'])
print(f"\nBest combination:")
print(f"Activation: {best_result['activation']}")
print(f"Architecture: {best_result['architecture']}")
print(f"MSE: {best_result['mse_mw']:.4f}")

# Create a more comprehensive plot
plt.figure(figsize=(12, 8))

# Group by activation function and plot each architecture
activations = results_df['activation'].unique()
architectures_list = results_df['architecture'].unique()

# Create subplot for each activation function
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for i, activation in enumerate(activations):
    subset = results_df[results_df['activation'] == activation]
    axes[i].bar(subset['architecture'], subset['mse_mw'], color=f'C{i}')
    axes[i].set_title(f'{activation} Activation Function')
    axes[i].set_xlabel('Architecture')
    axes[i].set_ylabel('MSE of Mw')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Also create a heatmap-style comparison
pivot_table = results_df.pivot(index='activation', columns='architecture', values='mse_mw')
plt.figure(figsize=(12, 6))
plt.imshow(pivot_table.values, cmap='RdYlBu_r', aspect='auto')
plt.colorbar(label='MSE of Mw')
plt.xticks(range(len(pivot_table.columns)), pivot_table.columns, rotation=45)
plt.yticks(range(len(pivot_table.index)), pivot_table.index)
plt.title('MSE Heatmap: Activation Functions vs Architectures')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import (
    explained_variance_score, max_error, mean_absolute_error,
    mean_squared_error, mean_squared_log_error, median_absolute_error,
    r2_score
)

# First, retrain the best model to get predictions
best_result = min(results, key=lambda x: x['mse_mw'])
print(f"Evaluating best model: {best_result['activation']} with {best_result['architecture']}")

# Get the best architecture configuration
best_activation = None
best_layers = None
for activation, tf_name in transfer_functions.items():
    if tf_name == best_result['activation']:
        best_activation = activation
        break

for arch_name, layers in architectures.items():
    if arch_name == best_result['architecture']:
        best_layers = layers
        break

# Rebuild and retrain the best model
best_model = Sequential()
best_model.add(Input(shape=(X_train.shape[1],)))

for neurons in best_layers:
    best_model.add(Dense(neurons, activation=best_activation))
    
best_model.add(Dense(2, activation='linear'))

best_model.compile(optimizer=Adam(), loss='mse')
best_model.fit(X_train, y_train,
               batch_size=32,
               epochs=25,
               validation_data=(X_val, y_val),
               verbose=0,
               shuffle=True)

# Get predictions from the best model
y_pred = best_model.predict(X_test)

# Create a dictionary of metrics that work for single-output regression
metrics = {
    "Explained Variance": explained_variance_score,
    "Max Error": max_error,
    "MAE": mean_absolute_error,
    "MSE": mean_squared_error,
    "MSLE": mean_squared_log_error,
    "Median AE": median_absolute_error,
    "R²": r2_score
}

# Loop through Mn (index 0) and Mw (index 1)
for i, target_name in enumerate(["Mn", "Mw"]):
    print(f"\n--- Metrics for {target_name} (Best Model: {best_result['activation']} - {best_result['architecture']}) ---")
    y_true = y_test[:, i]
    y_pred_single = y_pred[:, i]
    
    for metric_name, metric_func in metrics.items():
        try:
            value = metric_func(y_true, y_pred_single)
            print(f"{metric_name:20s}: {value:.4f}")
        except ValueError as e:
            print(f"{metric_name:20s}: Not supported ({e})")

# Also show comparison table
print(f"\n--- Actual vs Predicted for Best Model ---")
comparison = np.hstack((y_test, y_pred))
print("Actual Mn, Actual Mw, Predicted Mn, Predicted Mw:")
for i in range(min(10, len(comparison))):  # Show first 10 rows
    print(f"{comparison[i]}")
```
